```python
from lmfit import minimize, Parameters, Parameter, report_fit
from scipy.integrate import odeint
import copy
```

```python
def f0(x, t, ps):
    try:
        b = ps['b'].value
        tau = ps['tau'].value
    except:
        b, tau = ps
    
    I = x; 
    
    for i in np.arange(0, M+1, 1):
        I[i] = max(0, I[i]);
        
    dIdt = np.zeros(M+1); #one more species for phage
    
    dIdt[0] = - M/tau*I[0]; 
    for i in np.arange(1, M, 1):
        dIdt[i] = M/tau*(I[i-1] - I[i])
    dIdt[M] = M/tau*I[M-1]*b;
    
    return dIdt

def residual0(ps, ts, tdata0, ydata0, M):
    datalist = list(ydata0.keys());
    
    res_reweight = np.array([]); 
    x0 = np.zeros(M+1);
    x0[0] = cell_0; 
    x = odeint(f0, x0, ts, args=(ps,))
    
    for k in range(len(datalist)):
        time_raw = tdata0[k];
        data_real = ydata0[k];
    
        data_simul = np.zeros(len(time_raw)); 
        phage_tmp = np.sum(x, axis=1); 
        z = 0; 
        for j in time_raw:
            t_i = round(j);
            #print(t_i)
            data_simul[z] = (phage_tmp[t_i]-phage_tmp[0])/phage_tmp[0];
            z = z+1;
        res_tmp = (data_simul - data_real)**2;
#        res_tmp[3:] = res_tmp[3:]*100; 
        res_reweight = np.append(res_reweight, res_tmp);
    return res_reweight

```

```python
######################
cell_0 = 1e5;
ts = np.linspace(0,80,80)
params = Parameters()
params.add('b', value=300, min=10, max=1000)
params.add('tau', value=40, min=10, max=100)
######################

matplotlib.rcParams.update({"figure.figsize": (12,12)});
M_max = 16; 
res_all = np.zeros(M_max);
b_all = np.zeros(M_max);
tau_all = np.zeros(M_max);
phage_per_cell = np.zeros((M_max,80)); 

for M in np.arange(1,M_max+1,1):
    result = minimize(residual0, params, args=(ts,tdata,ydata,M), method='dual_annealing',nan_policy='omit'); 
    #result = minimize(residual0, params, args=(ts,t_data_sub,y_data_sub,M), method='dual_annealing',nan_policy='omit'); 
    
    res_all[M-1] = np.sum(result.residual);
    b_all[M-1] = result.params['b'].value; 
    tau_all[M-1] = result.params['tau'].value; 
    p0 = result.params; 
    x0 = np.zeros(M+1);
    x0[0] = cell_0; 
    x = odeint(f0, x0, ts, args=(p0,));
    all_phages = np.sum(x, axis=1); 
    phage_per_cell[M-1,:] = (all_phages)/all_phages[0];
    
    plt.subplot(4,4,M)
    plt.title('M='+str(M))
    for i in range(4):
        plt.plot(tdata[i], ydata[i],'o',alpha=0.2)
    
    plt.plot(t_data_all[0], y_data_all[0], 'o',color='grey',label='Data')
    plt.plot(ts, phage_per_cell[M-1,:],color='k',label='Model')
    plt.xlabel('Time (min)')
    plt.ylabel('PFU per inf')
plt.tight_layout()
```
